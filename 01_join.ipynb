## JOIN

```SQL
SELECT columns
FROM table1
INNER JOIN table2 ON
    table2.column_name = table1.column_name;
```

ตรง columns สามารถเป็นแบบ
```
table1.column, table2.column
```


หรือจะใช้ 
```sql
SELECT columns
FROM table1
INNER JOIN table2 USING(column)
```

#### INNER JOIN
- [link](https://www.sqlitetutorial.net/sqlite-inner-join/)

#### LEFT JOIN/LEFT OUTER JOIN
#### CROSS JOIN



````python
sql_command = """
SELECT albums.title, artists.name
FROM albums 
INNER JOIN artists ON artists.artistID = albums.artistId
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT albums.title, artists.name
FROM albums 
INNER JOIN artists USING(artistId);
"""

sql_query(sql_command, con)
````
