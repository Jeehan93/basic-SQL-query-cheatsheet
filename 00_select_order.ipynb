````python
import sqlite3
import pandas as pd
````

````python
con = sqlite3.connect("./chinook.db")
````

````python
con
````

````python
def sql_query(sql_command, con):
    return pd.read_sql_query(sql_command, con)
````

### SELECT
```sql
SELECT DISTINCT column_list
FROM table_list
  JOIN table ON join_condition
WHERE row_filter
ORDER BY column
LIMIT count OFFSET offset
GROUP BY column
HAVING group_filter;
```

````python
sql_command = """
SELECT *
FROM tracks 
LIMIT 5;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT TrackId, Name 
FROM tracks 
LIMIT 5;
"""

sql_query(sql_command, con)
````

### 
#### ตัวแปร ไม่ case sensitive

````python
sql_command = """
SELECT trackid, name 
FROM tracks 
LIMIT 10 OFFSET 10;
"""

sql_query(sql_command, con)
````

## 
### ORDER BY

```SQL
SELECT
   select_list
FROM
   table
ORDER BY
    column_1 ASC,
    column_2 DESC;
```

If you don’t specify the `ASC` or `DESC` keyword, SQLite sorts the result set using the `ASC` option.

````python
sql_command = """
SELECT TrackId, Name, Milliseconds, AlbumId
FROM tracks 
ORDER BY 
AlbumId ASC,
Milliseconds DESC;
"""

sql_query(sql_command, con)
````

### SQLite ORDER BY with the column position

```sql
SELECT TrackId, Name, Milliseconds, AlbumId
FROM tracks 
ORDER BY 
4 DESC, 3 ASC
```
- column position เริ่มนับที่ 1
- จะเห็นว่า `AlbumId` มี position = 4

````python
sql_command = """
SELECT TrackId, Name, Milliseconds, AlbumId
FROM tracks 
ORDER BY 
4 DESC, 3 ASC
"""

sql_query(sql_command, con)
````

## 
### เรื่องของ `NULL`

- `NULL` = ข้อมูลที่ unknown  ข้อมูลที่ missing
- ไม่สามรถเปรียบเทียบ `NULL` กับค่าอื่นได้
- `NULL = NULL` ได้ผลเป็น `FALSE`

> ปกติ SQLite แสดงผล NULL ขึ้นก่อน

- ใช้ `NULLS FIRST` หรือ `NULLS LAST` ใน ORDER BY ได้
- `IS NULL`, `IS NOT NULL`

````python
sql_query("SELECT 2 = 2;", con)
````

````python
sql_query("SELECT 1 = 2;", con)
````

````python
sql_query("SELECT NULL = NULL;", con)
````

````python
sql_command = """
SELECT trackId, name, composer
FROM tracks
ORDER BY
composer NULLS FIRST;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT trackId, name, composer
FROM tracks
WHERE
composer = NULL;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT trackId, name, composer
FROM tracks
WHERE
composer IS NULL
ORDER BY
name;
"""

sql_query(sql_command, con)
````

> เพราะว่า NULL ไม่สามารถเปรียบเทียบกันได้

### SELECT DISTINCT

```sql
SELECT DISTINCT select_list
FROM table;
```

- เลือก row ที่ไม่ duplicate จาก columns ใน select_list
- `DISTINCT` ต้องอยู่หลังคำว่า `SELECT`

````python
sql_command = """
SELECT DISTINCT composer
FROM tracks
ORDER BY
composer NULLS FIRST;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT city
FROM customers
ORDER BY city
LIMIT 15;
"""

sql_query(sql_command, con)
````

## 
### WHERE
```SQL
SELECT
    column_list
FROM
    table
WHERE
    search_condition;
```

#### comparison operator
`=`, `!=`, `<`, `>`, `<=`, `>=`


#### logical operator
- `AND`, `NOT`, `OR`
    - `NOT IN`, `NOT BETWEEN`
- `ALL`, `ANY`
- `BETWEEN`: returns 1 if a value is within a range.
    - `BETWEEN a AND b`
- `EXIST`: returns 1 if a subquery contains any rows.
- `IN`: returns 1 if a value is in a list of values.
- `LIKE`: returns 1 if a value matches a pattern

````python
sql_command = """
SELECT trackId, name, composer, milliseconds
FROM tracks
WHERE
milliseconds > 10000 AND
trackId < 300
ORDER BY
milliseconds;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT trackId, name, composer, milliseconds
FROM tracks
WHERE
milliseconds BETWEEN 10000 AND 30000
ORDER BY
milliseconds;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT trackId, name, composer, milliseconds
FROM tracks
WHERE
trackId IN (2020, 02, 17, 18)
"""

sql_query(sql_command, con)
````

## 
### IN, IN + subquery

```sql
SELECT *
FROM table
WHERE expression IN (value|subquery)
```

````python
sql_command = """
SELECT albumid
FROM albums
WHERE artistid = 12;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT trackId, name, composer, milliseconds, albumId
FROM tracks
WHERE
albumId IN (
    SELECT albumId
    FROM albums
    WHERE artistId = 12
)
"""

sql_query(sql_command, con)
````
