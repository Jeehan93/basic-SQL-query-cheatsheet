## CASE

The SQLite `CASE` expression to add the conditional logic to a query.

The `CASE` expression is similar to the `IF-THEN-ELSE` statement in other programming languages.

```sql
CASE case_expression
     WHEN when_expression_1 THEN result_1
     WHEN when_expression_2 THEN result_2
     ...
     [ ELSE result_else ] 
END
```

- ใช้สำหรับแปลงค่าเมื่อ query เช่น ช-ญ เก็บเป็น M-F แต่ต้องการแสดงเป็น Male-Female

````python
sql_command = """
SELECT customerid,
    firstname,
    lastname,
    CASE country 
        WHEN 'USA' THEN 'Domestic' 
        ELSE 'Foreign' 
    END CustomerGroup
FROM 
    customers
ORDER BY 
    lastname,
    firstname
LIMIT 10;
"""

sql_query(sql_command, con)
````

````python
sql_command = """
SELECT
    trackid,
    name,
    CASE
    WHEN milliseconds < 60000 THEN 'short'
    WHEN milliseconds > 60000 AND milliseconds < 300000 THEN 'medium'
    ELSE 'long'
    END category_duration
FROM tracks
LIMIT 20;
"""

sql_query(sql_command, con)
````
